```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_readings = 200
drexel_last4 = 1234  # REPLACE with your last 4 digits of your Drexel ID
rng = np.random.default_rng(drexel_last4)

# Generate timestamps (uniform) and sort them for monotonic time
timestamps = rng.uniform(0, 10, size=n_readings)
timestamps.sort()

# Generate sensor readings using normal distributions
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n_readings)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n_readings)

# Quick visualization
plt.figure(figsize=(8, 4))
plt.plot(timestamps, sensor_a, marker='o', linestyle='-', markersize=4, label='Sensor A (μ=25, σ=3)')
plt.plot(timestamps, sensor_b, marker='s', linestyle='--', markersize=4, label='Sensor B (μ=27, σ=4.5)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Simulated Temperature Readings (200 samples each)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of sensor data (vs time) and A vs B correlation
plt.figure(figsize=(8, 4))
plt.scatter(timestamps, sensor_a, s=30, alpha=0.7, label='Sensor A')
plt.scatter(timestamps, sensor_b, s=30, alpha=0.5, label='Sensor B')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Scatter: Temperature vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter: Sensor A vs Sensor B
plt.figure(figsize=(5,5))
plt.scatter(sensor_a, sensor_b, c='tab:purple', alpha=0.6)
plt.xlabel('Sensor A (°C)')
plt.ylabel('Sensor B (°C)')
plt.title('Sensor B vs Sensor A')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
# Implementation of the intent: scatter plot (Sensor vs Time)
plt.figure(figsize=(8,5))
plt.scatter(timestamps, sensor_a, c='tab:blue', label='Sensor A', s=40, alpha=0.8)
plt.scatter(timestamps, sensor_b, c='tab:orange', label='Sensor B', s=40, alpha=0.8)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Scatter: Temperature vs Time (Sensor A & B)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
# Implementation: overlaid histograms with means
plt.figure(figsize=(8,5))
bins = 30
plt.hist(sensor_a, bins=bins, alpha=0.5, color='tab:blue', label='Sensor A')
plt.hist(sensor_b, bins=bins, alpha=0.5, color='tab:orange', label='Sensor B')
mean_a = sensor_a.mean()
mean_b = sensor_b.mean()
plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label='Sensor A mean')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label='Sensor B mean')
plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Overlaid Histogram: Sensor Temperature Distributions')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
# Implementation: side-by-side box plot comparing Sensor A and Sensor B
plt.figure(figsize=(6,5))
data = [sensor_a, sensor_b]
labels = ['Sensor A', 'Sensor B']
box = plt.boxplot(data, labels=labels, patch_artist=True, widths=0.6)
# Color the boxes to match earlier plots
colors = ['tab:blue', 'tab:orange']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
# Emphasize medians
for median in box['medians']:
    median.set_color('black')
    median.set_linewidth(2)
# Overall mean across both sensors
overall_mean = np.concatenate((sensor_a, sensor_b)).mean()
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=1.5, label='Overall mean')
plt.ylabel('Temperature (deg C)')
plt.title('Box Plot: Sensor Temperature Comparison')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
